```python
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Open the image file with PIL
pil_img = Image.open(r'Yale Face Database\data\subject01.centerlight')  # replace with your image path

# Convert PIL Image to OpenCV Image (NumPy array)
img = np.array(pil_img)

# Convert grayscale image to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)

# Detect faces
faces = face_cascade.detectMultiScale(img, 1.1, 4)

# Draw rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(img_rgb, (x, y), (x+w, y+h), (255, 255, 0), 2)  # BGR color for the rectangle is now yellow

# Display the output
plt.imshow(img_rgb)
plt.show()
```
